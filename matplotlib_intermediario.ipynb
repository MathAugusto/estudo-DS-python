**Estudo sobre a biblioteca MatPlotLib**

```
# import do Matplotlib e do inline para poder plotar sem o comando plt.show()
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Criando um numpy array para servir de base para os gráficos de teste
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```
print(x)
```

```
print(y)
```

**Plots e Gráficos**

```
# Funcional
plt.plot(x,y)
plt.xlabel('label X')
plt.ylabel('label Y')
plt.title('TÍTULO')
```

```
# Subplots e modificação 
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(x,y,'b--')
```

```
# OO mod
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('Label X')
axes.set_ylabel('Label Y')
axes.set_title('TÍTULO')
```

```
# Fazendo um figura dentro da outra: axes(posicao_lateral, posicao_vertical_, lagura_, altura_)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Grande')

axes2.plot(y,x)
axes2.set_title('Pequeno')
```

```
# Multiplos plots em um mesmo tammaho de figura
fig, axes = plt.subplots(nrows=3, ncols=3,figsize=(8,5))
plt.tight_layout()
```

```
# Plotando determinada figura traves de seu index:
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
axes[0,0].plot(x,y)
axes[0,0].set_title('WWWW')

axes[1,1].plot(y,x)
axes[1,1].set_title('Hoi')
```

```
#colocando legendas
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(x,y,label='azul')
axes1.plot(y,x,label='laranja')
axes1.set_title('title')
#loc= é para selecionar onde ficaráá a legenda, usando 0 ele busca o "melhor" lugar, ou passando coordenadas loc=(0.1, 0.1) para posiconar
axes1.legend(loc=0)
```

```
# Zoom em gráficos
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(x,y)
axes1.plot(y,x)
axes1.set_title('title')

#selecionar o tamanho da visualização nos eixos X e Y:

axes1.set_xlim(0,5)
axes1.set_ylim(0,2)
```

```
# Melhorando o visual de gráficos
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y**2.3,label='azul', color='purple',lw=3,alpha=0.7,ls='-',marker='*',markersize=15,markerfacecolor='yellow')
axes.plot(x,y**2,label='laranja', color='red',lw=8,alpha=0.3,ls='-.')
```
