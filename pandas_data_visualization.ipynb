**DATA VISUALIZATION COM PANDAS**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df3 = pd.read_csv("df3")
plt.style.use("ggplot")
```

```
df3.info()
```

```
df3.head()
```

```
# Scatterplot
df3.plot.scatter(x="a", y="b", s=50, c="red", figsize=(12,3))
```

```
# Histogram
df3["a"].plot.hist(bins=20, alpha=0.5)
```

```
# Boxplot
df3[['a', 'b']].plot.box()
```

```
# KDEplot
df3['d'].plot.kde(ls="--", lw=4)
```

```
# Areaplot
f = plt.figure()
df3.iloc[0:30,:].plot.area(alpha=0.7)
# Para por a legenda fora da rea do grfico
plt.legend(loc="center left", bbox_to_anchor=(1.0, 0.5))
```
